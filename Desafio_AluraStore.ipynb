### Importação dos dados

```python
# Desafio Alura Store

import pandas as pd
import matplotlib.pyplot as plt

# URLs
urls = {
    "loja_1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "loja_2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "loja_3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "loja_4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
}
```

```python
# Carregar os dados
dados_lojas = {nome: pd.read_csv(url) for nome, url in urls.items()}
```

#1. Análise do faturamento

```python
faturamentos = {nome: df['Preço'].sum() for nome, df in dados_lojas.items()}
faturamentos
```

```python
# Faturamento total
plt.figure(figsize=(8, 5))
plt.bar(faturamentos.keys(), faturamentos.values(), color='skyblue')
plt.title('Faturamento Total por Loja')
plt.ylabel('R$')
plt.show()
```

# 2. Vendas por Categoria

```python
categorias_vendidas = {nome: df['Categoria do Produto'].value_counts() for nome, df in dados_lojas.items()}
categorias_vendidas
```

```python
for nome, df in dados_lojas.items():
    categorias = df['Categoria do Produto'].value_counts()
    plt.figure(figsize=(8, 5))
    categorias.plot(kind='barh', color='teal')
    plt.title(f'Vendas por Categoria - {nome}')
    plt.xlabel('Quantidade Vendida')
    plt.ylabel('Categoria')
    plt.tight_layout()
    plt.show()
```

# 3. Média de Avaliação das Lojas

```python
avaliacoes_medias = {nome: df['Avaliação da compra'].mean() for nome, df in dados_lojas.items()}
avaliacoes_medias
```

```python
# Avaliações médias
plt.figure(figsize=(8, 5))
plt.plot(avaliacoes_medias.keys(), avaliacoes_medias.values(), marker='o', linestyle='-', color='green')
plt.title('Média das Avaliações por Loja')
plt.ylabel('Nota Média')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
mais_menos_vendidos = {
    nome: {
        "mais_vendido": df['Produto'].value_counts().idxmax(),
        "menos_vendido": df['Produto'].value_counts().idxmin()
    }
    for nome, df in dados_lojas.items()
}
mais_menos_vendidos
```

```python
for nome, df in dados_lojas.items():
    contagem = df['Produto'].value_counts()
    mais_vendidos = contagem.head(5)
    menos_vendidos = contagem.tail(5)

    fig, ax = plt.subplots(1, 2, figsize=(12, 5))

    mais_vendidos.plot(kind='bar', ax=ax[0], color='green')
    ax[0].set_title(f'{nome} - Top 5 Mais Vendidos')
    ax[0].set_ylabel('Quantidade')

    menos_vendidos.plot(kind='bar', ax=ax[1], color='orange')
    ax[1].set_title(f'{nome} - Top 5 Menos Vendidos')

    plt.tight_layout()
    plt.show()
```

# 5. Frete Médio por Loja

```python
frete_medio = {nome: df['Frete'].mean() for nome, df in dados_lojas.items()}
frete_medio
```

```python
# Frete médio
plt.figure(figsize=(8, 5))
plt.scatter(frete_medio.keys(), frete_medio.values(), color='red')
plt.title('Custo Médio de Frete por Loja')
plt.ylabel('R$')
plt.show()
```
